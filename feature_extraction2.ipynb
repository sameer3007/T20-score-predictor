```python
import pandas as pd
import pickle
import numpy as np
```

```python
# Load CSV into a DataFrame
df = pd.read_csv("extracted_data.csv")

# Check first few rows
print(df.shape)
df.head()
```

## required input format
### batting-team
### bowling-team
### city
### current_score
### ball_left
### wickets_left
### current_rr

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df
```

```python
df.drop(columns=["venue"], inplace=True)
```

```python
df
```

```python
df['city'].value_counts()
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id')['runs_batsman'].cumsum()
```

```python
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball'].astype('int')
```

```python
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['wicket_player_out'] = df['wicket_player_out'].apply(lambda x:0 if x=='0' else 1)
df['wicket_player_out'] = df['wicket_player_out'].astype('int')
df['wicket_player_out'] = df.groupby('match_id')['wicket_player_out'].cumsum()
df['wickets_left'] = 10 - df['wicket_player_out']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs_batsman'].sum().values.tolist())
```

```python
last_five
```

```python
df['last_five_runs'] = last_five
```

```python
df
```

```python
df['crr']=round(df['crr'],2)
```

```python
df
```

```python
df['total_runs'] = df.groupby('match_id')['runs_batsman'].transform('sum')
```

```python
df
```

```python
final_df=df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr',
             'last_five_runs',"total_runs"]].copy()
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df
```

```python
#shuffle the df to remove any ordering that can lead to bias model
final_df = final_df.sample(frac=1).reset_index(drop=True)
```

```python
final_df
```

# model trainning and testing

```python
X = final_df.drop(columns=['total_runs'])
y = final_df['total_runs']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
# ColumnTransformer lets us apply different preprocessing steps to different columns in one go.
# Here, we will one-hot encode specific categorical columns and keep all other columns as they are.

trf = ColumnTransformer(
    transformers=[
        (
            'trf',  # Step name (can be any string, just an identifier for this transformation)
            
            # OneHotEncoder converts categorical variables into dummy variables (0/1 columns)
            # sparse_output=False → returns a normal NumPy array instead of a sparse matrix
            # drop='first' → drops the first category to avoid the dummy variable trap
            OneHotEncoder(sparse_output=False, drop='first'),
            
            # List of columns to apply the OneHotEncoder to
            ['batting_team', 'bowling_team', 'city']
        )
    ],
    
    # remainder='passthrough' means:
    # Keep all other columns in the dataset as they are (unchanged)
    remainder='passthrough',
    force_int_remainder_cols=False  # 👈 suppresses the warning and uses future behavior now
)

# How to use:
# trf.fit_transform(df) → will one-hot encode the given columns
# and return a NumPy array containing both the encoded and untouched columns.
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor

# Create a Pipeline that will:
# Step 1 → One-hot encode categorical columns using 'trf'
# Step 2 → Scale all numeric features
# Step 3 → Train an XGBoost regression model

pipe = Pipeline(steps=[
    ('step1', trf),  # Apply ColumnTransformer 'trf' (one-hot encode categorical columns, pass others through)
    
    ('step2', StandardScaler()),  # Standardize features (mean=0, std=1) — mainly for scale-sensitive models
    
    ('step3', XGBRegressor(
        n_estimators=1000,   # Number of boosting rounds (trees)
        learning_rate=0.2,   # How much each tree contributes (lower → slower but more accurate learning)
        max_depth=12,        # Maximum depth of each decision tree (controls complexity)
        random_state=1       # Seed for reproducibility
    ))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
#saving model
pickle.dump(pipe, open('xgboost_model.pkl', 'wb'))
```
